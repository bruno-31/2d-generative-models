```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```

```python
theta = np.random.uniform(low=2,high=5,size=300)
rho = 0 + 2 * theta + np.random.randn(300)*0.3
```

```python
x,y = pol2cart(rho,theta)
x2,y2 = pol2cart(rho,theta+1*np.pi/2)
x3,y3 = pol2cart(rho,theta+2*np.pi/2)
x4,y4 = pol2cart(rho,theta+3*np.pi/2)
# x5,y5 = pol2cart(rho,theta+4*np.pi/7)
# x6,y6 = pol2cart(rho,theta+5*np.pi/7)
```

```python
plt.scatter(x,y,s=5)
plt.scatter(x2,y2,s=5)
plt.scatter(x3,y3,s=5)
plt.scatter(x4,y4,s=5)
# plt.scatter(x5,y5,s=5)
# plt.scatter(x6,y6,s=5)
# # plt.scatter(x3,y3,s=5)

plt.show()
```
