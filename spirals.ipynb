```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```

```python
theta = np.random.uniform(low=2,high=5,size=300)
rho = 0 + 2 * theta + np.random.randn(300)*0.3
```

```python
x,y = pol2cart(rho,theta)
x2,y2 = pol2cart(rho,theta+1*np.pi/2)
x3,y3 = pol2cart(rho,theta+2*np.pi/2)
x4,y4 = pol2cart(rho,theta+3*np.pi/2)
# x5,y5 = pol2cart(rho,theta+4*np.pi/7)
# x6,y6 = pol2cart(rho,theta+5*np.pi/7)
```

```python
def spiral():
    def pol2cart(rho, phi):
        x = rho * np.cos(phi)
        y = rho * np.sin(phi)
        return(x, y)
    Y=[]
    X = []
    for i in range(4):
        theta = np.random.uniform(low=2,high=5,size=300)
        rho = 0 + 2 * theta + np.random.randn(300)*0.3
        x,y = pol2cart(rho,theta+i*np.pi/2)
        A=np.stack((x,y))
        X.append(A.T)
        lbl = np.ones(300)*i
        Y.append(lbl)
    return np.concatenate(X), np.concatenate(Y)
    
```

```python
x,y = spiral()
```

```python
def scatter_dataset(ax,X,y,s=10):
    for i in range(int(np.max(y))+1):
        ax.scatter(X[y==i][:,0],X[y==i][:,1],s=s)
        ax.scatter(X[y==i][:,0],X[y==i][:,1],s=s)
```

```python
f, ax = plt.subplots(1)
scatter_dataset(ax,x,y)
plt.show()
```
